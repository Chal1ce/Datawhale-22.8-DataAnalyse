```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

f,ax=plt.subplots(1,2,figsize=(20,10))

train[train['是否幸存']==0].年龄.plot.hist(ax=ax[0],bins=20,edgecolor='black',color='red')
ax[0].set_title('Died')
x1=list(range(0,85,5))
ax[0].set_xticks(x1)

train[train['是否幸存']==1].年龄.plot.hist(ax=ax[1],color='green',bins=20,edgecolor='black')
ax[1].set_title('Survived')
x2=list(range(0,85,5))
ax[1].set_xticks(x2)
plt.show()
```

```python
f,ax=plt.subplots(1,3,figsize=(20,8))
sns.distplot(train[train['乘客等级(1/2/3等舱位)']==1].是否幸存,ax=ax[0])
ax[0].set_title('一等舱票价')
sns.distplot(train[train['乘客等级(1/2/3等舱位)']==2].是否幸存,ax=ax[1])
ax[1].set_title('二等舱票价')
sns.distplot(train[train['乘客等级(1/2/3等舱位)']==3].是否幸存,ax=ax[2])
ax[2].set_title('三等舱票价')
plt.show()
```
